```python
import numpy as np
import pandas as pd
data=pd.read_excel('D:\M\MATLAB Driver\\forward\Fifty-forth\液相.xlsx',sheet_name='Sheet2')
data
```

```python
#将‘1.95±0.18’这种数据计算，第一行为1.95-0.18，第二行为1.95+0.18
#将data转换为numpy数组
data=np.array(data)
#删除第一列
data=np.delete(data,0,axis=1)
data_pocess=np.zeros((data.shape[0],4))
for i in range(25):
    if i<14:
        i=i*2
        for  j in range(data.shape[1]):
                temp=data[i,j].split('±')
                data_pocess[i,j]=float(temp[0])-float(temp[1])
                data_pocess[i+1,j]=float(temp[0])+float(temp[1])
    elif i==14:
        #只需要data的数据前半部分
        data_pocess[i*2,:4]=0
    else:
        i=(i-1)*2+1
        for  j in range(data.shape[1]):
            temp=data[i,j].split('±')
            data_pocess[i,j]=float(temp[0])-float(temp[1])
            data_pocess[i+1,j]=float(temp[0])+float(temp[1])
        
data_pocess
        
    
```

```python
#将‘1.95±0.18’这种数据计算，第一行为1.95-0.18，第二行为1.95+0.18
#将data转换为numpy数组
data=np.array(data)
#删除第一列
data=np.delete(data,0,axis=1)
data_pocess=np.zeros((data.shape[0],4))
for i in range(25):
    i=i*2
    for  j in range(data.shape[1]):
        temp=data[i,j].split('±')
        data_pocess[i,j]=float(temp[0])-float(temp[1])
        data_pocess[i+1,j]=float(temp[0])+float(temp[1])
    
data_pocess
        
    
```

```python
#将data所有1.95±0.18这种数据只保留1.95
data=data.replace('±.*','',regex=True)
data
```

```python
#单因素差异性分析
import scipy.stats as stats
#将data转换为numpy数组
data=np.array(data)
#将data转换为float类型
data=data.astype(float)
#前四列与后四列分别进行单因素差异性分析
result=np.zeros((data.shape[0],2))
for i in range(data.shape[0]):
    result[i,0],result[i,1]=stats.ttest_ind(data[i,:4],data[i,4:],equal_var=False)
    
result=pd.DataFrame(result,columns=['t值','p值'])
result
```

```python
result.to_excel('D:\M\MATLAB Driver\\forward\Fifty-forth\液相单因素差异性分析.xlsx',index=False)
```

```python
#对data的25行数据进行显著性分析
import scipy.stats as stats
result=np.zeros((data.shape[0],2))
for i in range(data.shape[1]):
    result[i,0],result[i,1]=stats.ttest_1samp(data[:,i],0)
```
